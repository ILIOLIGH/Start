## Таблица с заработной платой

Выбранные виды экономической деятельности: "Добыча полезных ископаемых" и "Рыболовство, рыбоводство". 
Данные отрасли были выбраны в связи с тем, что в них наблюдается достаточно высокий уровени заработных плат, но при этом сами отрасли между собой не особо взаимосвязаны.

Мной было проведено сравнение "Постановления Госстандарта России от 06.11.2001" и "ОК 029-2014 (КДЕС Ред. 2). Общероссийский классификатор видов экономической деятельности", определяющие классификацию экономической деятельности с 2001 до 2017 год и начиная с 2017 года, соответственно. Под добычей полезных ископаемых и рыболовством, рыбоводством в данных нормативно-правовых документах понимается одно и то же, в связи с чем я буду анализировать данные из разных этапов совместно.

```python
wages = pd.ExcelFile('wages.xlsx')
```

```python
wages1 = wages.parse('Part1')
wages1 = wages1.set_index('Unnamed: 0')
wages1
```

```python
fish_part1 = wages1.loc[['Рыболовство, рыбоводство']]
fish_part1 = fish_part1.values[0].astype(float).tolist()
fish_part1
```

```python
soil_part1 = wages1.loc[['Добыча полезных ископаемых']]
soil_part1 = soil_part1.values[0].astype(float).tolist()
soil_part1
```

```python
wages2 = wages.parse('Part2')
wages2 = wages2.set_index('Unnamed: 0')
wages2
```

```python
wages2.index
```

### Можно было бы почистить названия от пробелов с использованием цикла и строкового метода strip(). Но поскольку названий мало, в данном случае быстрее просто скопировать название как оно есть

```python
fish_part2 = wages2.loc[['    рыболовство и рыбоводство']]
fish_part2 = fish_part2.values[0].astype(float).tolist()
fish_part2
```

```python
soil_part2 = wages2.loc[['  добыча полезных ископаемых']]
soil_part2 = soil_part2.values[0].astype(float).tolist()
soil_part2
```

### Теперь сделаю таблицу, в которой содержатся все нужные данные по выбранным отраслям

```python
fish_part1.extend(fish_part2)
```

```python
fish = fish_part1
fish
```

```python
soil_part1.extend(soil_part2)
```

```python
soil = soil_part1
soil
```

```python
year = [i for i in range(2000, 2024)]
```

```python
year
```

```python
len(year) == len(soil) == len(fish)
```

```python
table = pd.DataFrame({
    'year': year, 
    'soil': soil,
    'fish': fish
    })
```

```python
table = table.merge(infl, how = 'left', left_on = 'year', right_on = 'Year')
```

```python
table = table.rename(columns={'Year': 'soil_real', 'Inflation': 'inflation'})
# для того, чтобы работал цикл, заполняющий данные по реальной заработной плате, 
# нам нужны уже созданные столбцы
```

```python
table.insert(4, 'fish_real', np.nan)
```

### Переведём всё в цены 2000 года. Для этого нам нужно убрать влияние инфляции.

```python
#Поскольку далее выскакивают предупреждения, занимающие много места, уберём эти предупреждения
import warnings
warnings.filterwarnings('ignore')
```

```python
table['soil_real'][0] = table['soil'][0]
```

```python
denominator = 1
for i in range(1, 24):
    denominator *= (1+table['inflation'][i]/100)
    table['soil_real'][i] = table['soil'][i]/denominator
```

```python
table['fish_real'][0] = table['fish'][0]
```

```python
denominator = 1
for i in range(1, 24):
    denominator *= (1+table['inflation'][i]/100)
    table['fish_real'][i] = table['fish'][i]/denominator
```

### Графики изменения средней зарплаты по годам для выбранных видов деятельности

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(table['year'], table['soil'], marker = '.', label = 'Добыча полезных ископаемых')
ax.plot(table['year'], table['fish'], marker = '.', label = 'Рыболовство и рыбоводство')
ax.set_xlabel('Год')
ax.set_ylabel('Средняя заработная плата')
ax.set_title('Среднемесячная номинальная заработная плата работников (2000-2023 гг.)')
ax.legend()
plt.show()
```

Как и предполагалось после визуальной оценки данных в таблице, график показал, что номинальные заработные платы были достаточно высокими для обеих отраслей. Тем не менее вплоть до 2019 года люди, работающие в отрасли "Добыча полезных ископаемых", получали больше денег. Но, начиная с 2019 по 2021, уровень з. п. практически сравнялся. В 2022 году и далее больше получают люди, занимающиеся "Рыболовством и рыбоводством".

### Динамика изменения реальных зарплат с учетом инфляции

```python
fig, ax1 = plt.subplots()
ax1.plot(table['year'], table['soil'], marker = '.', label = 'Номинальная з. п.')
ax1.plot(table['year'], table['soil_real'], marker = '.', label = 'Реальная з. п.')
ax1.set_xlabel('Год')
ax1.set_ylabel('Средняя заработная плата')
ax1.set_title('Сравнение номинальной и реальной з. п. в отрасли "Добыча полезных ископаемых" (2000-2023 гг.)')
ax1.legend()
plt.show()
```

```python
fig, ax2 = plt.subplots()
ax2.plot(table['year'], table['fish'], marker = '.', label = 'Номинальная з. п.')
ax2.plot(table['year'], table['fish_real'], marker = '.', label = 'Реальная з. п.')
ax2.set_xlabel('Год')
ax2.set_ylabel('Средняя заработная плата')
ax2.set_title('Сравнение номинальной и реальной з. п. в отрасли "Рыболовство и рыбоводство" (2000-2023 гг.)')
ax2.legend()
plt.show()
```

Как видно по обоим графикам, инфляция значительно ускоряет темпы роста номинальной средней заработной платы по сравнению с реальной. Для отрасли "Добыча полезных ископаемых" номинальная заработная плата растёт скорее линейно, а для "Рыболовства и рыбоводства" практически экспоненциально. 

```python
fig, ax3 = plt.subplots()
ax3.plot(table['year'], table['soil_real'], marker = '.', label = 'Добыча полезных ископаемых')
ax3.plot(table['year'], table['fish_real'], marker = '.', label = 'Рыболовство и рыбоводство')
ax3.set_xlabel('Год')
ax3.set_ylabel('Средняя заработная плата')
ax3.set_title('Среднемесячная реальная заработная плата работников (2000-2023 гг.)')
ax3.legend()
plt.show()
```

Реальная заработная плата также растёт, свидетельствуя о том, что зарплата поднималась не только с целью перекрыть влияние инфляции, но повышалось и реальное благополучие людей, занятых в этих отраслях. И снова заметим, что реальная з.п. в отрасли "Рыболовство и рыбоводство" росла более интенсивными темпами.

```python
fig, (ax, ax3) = plt.subplots(2, sharex = True)
ax.plot(table['year'], table['soil'], marker = '.', label = 'Добыча полезных ископаемых')
ax.plot(table['year'], table['fish'], marker = '.', label = 'Рыболовство и рыбоводство')
#ax.set_xlabel('Год')
ax.set_ylabel('Средняя з. п.')
ax.set_title('Среднемесячная номинальная заработная плата работников (2000-2023 гг.)')
ax.legend()
ax3.plot(table['year'], table['soil_real'], marker = '.', label = 'Добыча полезных ископаемых')
ax3.plot(table['year'], table['fish_real'], marker = '.', label = 'Рыболовство и рыбоводство')
ax3.set_xlabel('Год')
ax3.set_ylabel('Средняя з. п.')
ax3.set_title('Среднемесячная реальная заработная плата работников (2000-2023 гг.)')
ax3.legend()
plt.show()
```

На графике с реальной средней заработной платой мы можем наблюдать локальные минимумы и максимумы, а также временами не такой стремительный рост как на графике с номинальной средней заработной платой. Очевидно, это связано с тем, что работадатели вероятнее всего думают не о реальном благополучии работников, которое как раз и зависит от инфляции, а о том, чтобы постепенно индексировать заработную плату, чтобы она была больше, чем в прошлом году. То есть они ведут наблюдение за номинальными показателями.

Важно отметить то, о чём уже упоминалось ранее: средняя заработная плата в отрасли "Рыболовство и рыбоводство" росла более интенсивно, чем в отрасли "Добыча полезных ископаемых". Это довольно интересная динамика. Традиционно для России отрасль, связанная с добычей полезных ископаемых являлась очень важной: туда входит добыча нефти и газа, которые являются основными продуктами российского экспорта. И тем не менее рыболовство и рыбоводство догоняет и даже перегоняет эту отрасль по уровню заработной платы, что свидетельствует о её растущей важности.
